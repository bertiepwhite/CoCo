```python
import os
Lightcurves_folder = os.path.join(os.environ['PYCOCO_DATA_DIR'], 'lc')
import pandas as pd

SN_name = "SN1998S"
```

```python
file_path = os.path.join(Lightcurves_folder, (SN_name + ".dat"))
with open(file_path, 'r') as file:
    lines = file.readlines()
    lines[0] = lines[0].replace("# ", "")
with open(file_path, 'w+') as file:
    file.writelines(lines)
new = pd.read_csv(file_path, sep = " ")
new = new.sort_values( by=['filter', 'flux'], ascending=[True, True])
```

```python
file_path_old = os.path.join(Lightcurves_folder, (SN_name + "_stored.dat"))
with open(file_path_old, 'r') as file:
    lines = file.readlines()
    lines[0] = lines[0].replace("# ", "")
with open(file_path_old, 'w+') as file:
    file.writelines(lines)
old = pd.read_csv(file_path_old, sep = " ")
old = old.sort_values( by=['filter', 'flux'], ascending=[True, True])
```

```python
old. drop("flux_err", axis = 1)
```

```python
old['flux_err'] = new['flux_err']
```

```python
old['flux_err'] = pd.Series(["{0:.6g}".format(val) for val in old['flux_err']], index = old.index)
old['flux'] = pd.Series(["{0:.6g}".format(val) for val in old['flux']], index = old.index)
```

```python
old.columns = ['MJD', "flux", "flux_err", "filter"]
```

```python
os.remove(path=file_path)
old.to_csv(path_or_buf=file_path, sep = " ", index = False)
with open(file_path, 'r') as file:
    lines = file.readlines()
    lines[0] = "# " + lines[0]
with open(file_path, 'w+') as file:
    file.writelines(lines)
```

```python
old = old.sort_values( by=['filter', 'flux'], ascending=[True, True])
```
